# Figure S3

Code to generate Figure S3 of Nyquist et al 2022

```python
import scanpy as sc
import matplotlib.pyplot as plt
```

```python
S3_path = "../../Results/plots/figure_S3/"
```

```python
adata_no_doublets=sc.read_h5ad("../../Data/processed_data/all_timepoints/adata_no_doublets_FINAL.h5ad")
```

```python
plt.rcParams["figure.figsize"]=(4,4)
# somatic markers from twigger et al 
markers = {"VA":["SPARCL1", "APOE", "PDK4","ANGPT2","TFPI","AKAP12","GJA4","MT1A"], #VA
           "EN":["CDH5","CD93","CLDN5","ECSCR","ENG","S1PR1","SELE"], # EN
          "FB":["DCN","TNFAIP6","MMP2","LUM","PDGFRA","COL1A1","IGFBP6","SERPINE2"]} #FB]
sc.pl.dotplot(adata_no_doublets,var_names=markers,groupby="simplified celltypes",standard_scale="var",colorbar_title='Standardized mean\nexpression in group',show=False)
plt.savefig(S3_path + "/nonimmune_marker_genes_twigger.pdf",bbox_inches="tight")
```

```python
epi_markers = {"MY":["ACTA2","MME","TP63","TAGLN","MYLK","OXTR","PROCR","KRT5","KRT14","KRT17"],
              "Luminal":["EPCAM","KRT18","KRT19","ELF3","CLDN3","CLDN4","GATA3","ITGA6"],
              "HR":['PGR',"ESR1","PRLR","AREG","AGR2","ANKRD30A","FOXA1","TFF1","TFF3"],
              "LP":["ALDH1A3","SLPI","ELF3","KRT23","KIT","MMP7","PROM1","KRT15","BCL11A","ALDH1A3"],
              "LC-both":["LALBA","CSN2","CSN3","ELF5","LTF","S100A9","CD36","XDH","MUC1","SPP1"],
              "LC1":["MT1E","JUN","SLC30A1","MT1F","MT1G","MT1M","VEGFA","IL6R","NEAT1"],
              "LC2":["LY6D"]}

sc.pl.dotplot(adata_no_doublets,var_names=epi_markers,groupby="simplified celltypes",standard_scale="var",colorbar_title='Standardized mean\nexpression in group',show=False)
plt.savefig(S3_path + "/epi_marker_genes_major_celltypes_twigger.pdf",bbox_inches="tight")
```

```python
sc.pl.dotplot(adata_no_doublets[~adata_no_doublets.obs['Epithelial Cell Subclusters'].isin(["nan"])],var_names=epi_markers,groupby='Epithelial Cell Subclusters',standard_scale="var",colorbar_title='Standardized mean\nexpression in group',show=False)
plt.savefig(S3_path + "/epi_marker_genes_subclusters_twigger.pdf",bbox_inches="tight")
```

```python
#murrow markers

murrow_markers = {"Basal/myoepithelial":["ACTA2","ACTG2","KRT14","KRT17","KRT5","MYL9","MYLK","SFN"],
                 "luminal-both":["ELF3","EPCAM","KRT18","KRT19","KRT8","MUC1"],
                 "HR+ luminal":["AGR2","ALCAM","ANKRD30A","AR","AREG","PIP","TFF1","TFF3"],
                  "Secretory luminal":["ALDH1A3","BBOX1","GABRP","KIT","KRT15","LTF","PROM1","SLPI"],
                 "Fibroblast":["COL1A1","DCN","FN1","IGFBP6","MMP2","PDGFRA","PDPN"],
                 "endothelial":["ESAM","ANGPT2","CDH5","CLDN5","ECSCR","FABP4","ENG","PLVAP"],
                 "vascular accessory":["CPM","ENPEP","GJA4","KCNE4","PLN"]}

sc.pl.dotplot(adata_no_doublets,var_names=murrow_markers,groupby="simplified celltypes",standard_scale="var",colorbar_title='Standardized mean\nexpression in group',show=False)
plt.savefig(S3_path + "/major_marker_genes_major_celltypes_murrow.pdf",bbox_inches="tight")
```

```python
sc.pl.dotplot(adata_no_doublets[~adata_no_doublets.obs['Epithelial Cell Subclusters'].isin(["nan"])],var_names=murrow_markers,groupby='Epithelial Cell Subclusters',standard_scale="var",colorbar_title='Standardized mean\nexpression in group',show=False)
plt.savefig(S3_path + "/major_marker_genes_subsets_murrow.pdf",bbox_inches="tight")
```
